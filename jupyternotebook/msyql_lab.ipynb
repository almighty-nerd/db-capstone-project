```python
import mysql.connector as connector

connection = connector.connect(user="root", password="root",db="mydb") 

cursor = connection.cursor()




show_tables_query = "SHOW tables" 
cursor.execute(show_tables_query)

# Fetching all results from the executed query
results = cursor.fetchall()

print(results)
```

```python
query = "SELECT * FROM customers"  
cursor.execute(query)

# Fetching all results from the executed query
results = cursor.fetchall()

# Displaying the results
for row in results:
    print(row)
```

```python
jointable = """
SELECT o.orderid,c.customername,l.city, l.country, SUM(s.cost) AS total_order_cost
FROM Customers c
INNER JOIN Orders o ON c.customerid = o.customerid
INNER JOIN Locations l on c.locationid = l.locationid
INNER JOIN Orderdetails od on o.orderid = od.orderid
INNER JOIN Sales s on od.orderdetailid = s.orderdetailid
WHERE s.cost > 60
GROUP BY o.orderid
;

"""
cursor.execute(jointable)
results = cursor.fetchall()
for row in results:
    print(row)
```
